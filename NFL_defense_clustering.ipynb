Imports

```python
import os
import time
import umap

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from DCBV import DBCV

from scipy.interpolate import griddata
from sklearn.cluster import DBSCAN, KMeans, SpectralClustering
from sklearn.manifold import Isomap
from sklearn.decomposition import PCA
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import normalize
from sklearn.utils.graph_shortest_path import graph_shortest_path
from sklearn.metrics import silhouette_score

from plot_play import plot_play
```

Create function to combine plays from dataset. We are only keeping player positions from the moment the ball is snapped for each play.

```python
def combine_games(path, event):
    temps = []
    for fname in os.listdir(path):
        temp = pd.read_csv(path+fname, index_col=[0])
        temp = temp.loc[temp['event'] == event]
        temps.append(temp)
    df = pd.concat(temps)
    return df
```

Import player data, play-by-play data, and ball-snap player location data.

```python
#read data
comb = combine_games('Data/Tracking/', 'ball_snap')
pbp = pd.read_csv('Data/play_by_play_2017.csv')
players = pd.read_csv('Data/players.csv', index_col=[0])
plays = pd.read_csv('Data/plays.csv', index_col=[0])
```

Create combined dataframe with player, play-by-play, and location data. Remove any special teams plays or penalties from data.

```python
comb = comb.merge(players[['nflId', 'PositionAbbr', 'FirstName', 'LastName']], on='nflId', how='inner')
comb = comb.merge(plays[['gameId', 'playId', 'isSTPlay', 'isPenalty']], on=['gameId', 'playId'], how='inner')
comb = comb.loc[(comb['isSTPlay']==False)&(comb['isPenalty']==False)]
#comb = comb.merge(pbp[['old_game_id', 'play_id', 'penalty_type']], left_on=['gameId', 'playId'], right_on=['old_game_id', 'play_id'], how='left')
#print(comb.shape)
#comb = comb.loc[~(comb['penalty_type'].isin(['Defensive Offsides', 'False Start']))]
print(comb.columns)
```

To determine the location of the ball, we use the position of the center at the moment of the snap. The center touches the ball to start each play.

```python
#use the center's position as a proxy for the ball location at snap
ball = comb.loc[comb['PositionAbbr'] == 'C']
ball = ball[['x', 'y', 'gameId', 'playId', 'frame.id', 'team']]
ball.rename(columns={'x': 'x_ball', 'y': 'y_ball', 'team': 'offense'}, inplace=True)
comb = comb.merge(ball, on = ['gameId', 'playId', 'frame.id'], how='inner')
```

Subtract the ball position coordinates from the player location coordinates, and flip all coordinates so they are positive, standardizing the player locations for each snap. 

```python
#find each player's position relative to the ball at snap
comb = comb.loc[~(comb['team']==comb['offense'])]
comb = comb.groupby(['gameId', 'playId','nflId']).first().reset_index()

comb['x_diff'] = comb['x']-comb['x_ball']
comb['y_diff'] = comb['y']-comb['y_ball']

#flip player position relative to ball so all distances are on the same axis
rev = np.sign(comb['x_diff']) 
comb['x_diff'] = comb['x_diff']*rev
comb['y_diff'] = comb['y_diff']*rev
```

Demo plot of player starting positions for a specific play

```python
demo = comb.loc[(comb['gameId']==2017090700)&(comb['playId']==94)]
plt.scatter(demo['x_diff'], demo['y_diff']);
```

Outside linebackers 'OLB' often have different starting locations and responsibilities based on scheme and individual skillset. Some players labeled as 'OLB' in a 3-4 defense have roles more similar to a traditional 'DE'. To accurately identify the roles of players labeled as an 'OLB', I manually divide players based on their average starting depth relative to the line of scrimmage. I used histogram of the average starting depth for all players labeled as outside linebackers to manually identify a cutoff of 2 yards. Those below the cutoff are relabeled as 'DE', those above remain 'OLB'. 

```python
#designate true OLB position based on average depth at snap
OLB = comb.loc[comb['PositionAbbr']=='OLB']
OLB = OLB.groupby('nflId').agg({'x_diff': 'mean', 'FirstName': 'max', 'LastName': 'max'})
plt.hist(OLB['x_diff'], bins = 20)
plt.title('Avg. starting depth from LOS')

cutoff = 2

plt.axvline(cutoff, c='red')
OLB['Pos'] = 0
OLB.loc[OLB['x_diff']>cutoff, 'Pos'] = 'OLB'
OLB.loc[OLB['x_diff']<=cutoff, 'Pos'] = 'DE'
```

Create table to display DE/OLB cutoff. Players selected above and below seem to pass the eye test w/r/t correct position labeling.

```python
tab = OLB.loc[(OLB['x_diff']>1.75)&(OLB['x_diff']<3)].copy()
tab.sort_values(by='x_diff', inplace=True)
tab['Average Depth'] = np.round(tab['x_diff'], 2)
tab.drop(columns=['x_diff'], inplace=True)
plt.table(cellText=tab.values, colLabels=tab.columns, loc='upper center')
plt.axis('off')
plt.axhline(0.535, c='red')
plt.show()
#plt.savefig('Figures/OLBcutoff.png')
```

Update combined dataframe with new position labels.

```python
OLB.reset_index(inplace=True)
comb = comb.merge(OLB[['nflId', 'Pos']], on='nflId', how='left')
comb['Pos'].fillna('', inplace=True)
comb['PositionAbbr'] = comb.apply(lambda x: x['PositionAbbr'] if not x['Pos'] else x['Pos'], axis=1)
```

Use position labels to seperate players into defensive line, linebacker, and defensive back categories 

```python
#assign dummy variables for postion groups
lb = ['ILB', 'OLB', 'MLB', 'FB', 'LB']
db = ['SS', 'FS', 'CB', 'TE', 'WR', 'DB']
dl = ['DE', 'DT', 'NT', 'G']

comb['DL'] = 0
comb.loc[comb['PositionAbbr'].isin(dl), 'DL'] = 1
comb['LB'] = 0
comb.loc[comb['PositionAbbr'].isin(lb), 'LB'] = 1
comb['DB'] = 0
comb.loc[comb['PositionAbbr'].isin(db), 'DB'] = 1

comb['Pos'] = comb.apply(lambda x: np.array((x['DL'], x['LB'], x['DB'])), axis=1)
```

Group dataframe by play ID. All player starting positions for a given play are stored as a list.

```python
#create separate df to calculate location data
location = comb[['gameId', 'playId', 'x_diff', 'y_diff', 'PositionAbbr', 'Pos']].copy()
location = location.groupby(['gameId', 'playId']).agg(\
    {'x_diff': lambda x: list(x), 'y_diff': lambda x: list(x), 'Pos': lambda x: list(x), 'PositionAbbr': 'count'}).reset_index()
location = location.loc[location['PositionAbbr']==11]
```

Create 3-digit integer representation of defensive personnel, with format #DL#LB#DB

```python
pos  = np.vstack(np.sum([np.array(i) for i in location['Pos']], axis=1))
mult=np.array([100,10,1])
pos = pos@mult
location['Personnel'] = pos
```

```python
#demo of personnel identifying integers
print(pos)
```

Create 2d histograms to quantize player positions into roughly 1-yard x 1-yard buckets. h is the number of bins for each dimension.

```python
h=50

y_max = comb['y_diff'].max()
y_min = comb['y_diff'].min()
x_max = comb['x_diff'].max()
x_min = comb['x_diff'].min()
xs = location['x_diff'].tolist()
ys = location['y_diff'].tolist()

start = time.time()
hists = []
for i in range(location.shape[0]):
    H, xedges, yedges = np.histogram2d(xs[i], ys[i], bins=[h,h], range=[[x_min, x_max], [y_min, y_max]])
    hists.append(H.ravel())
    
end = time.time()
print(end - start)
```

Calculate silhouette score over range of DBSCAN hyperparameters and select hyperparameters that maximize silhouette score.

```python
e = []
s = []
cv = []
for eps in np.arange(0.1, 0.2, 0.01):
    for samps in range(15, 50, 5):
        np.random.seed(47)
        scan = DBSCAN(eps=eps, min_samples=samps).fit(u)
        e.append(eps)
        s.append(samps)
        cv.append(silhouette_score(u, scan.labels_))
idx = np.argmax(cv)
print(e[idx], s[idx])
```

Create heatmap of silhouette score

```python
def plot_sil_score(e, s, cv):
    e=np.array(e)
    s=np.array(s)
    cv = np.array(cv)
    grid_x, grid_y = np.mgrid[e.min():e.max():1000j, s.min():s.max():1000j]
    grid = griddata((e,s), cv, (grid_x, grid_y), method='cubic')
    fig, ax = plt.subplots(1,1,figsize=(12,8))
    im = ax.imshow(grid.T, extent=[e.min(),e.max(),s.min(),s.max()], origin='lower', aspect='auto', cmap='plasma')
    plt.xlabel('Epsilon')
    plt.ylabel('Minimum Samples per Cluster')
    plt.title('Silhouette Score Of DBSCAN Cluster Assignments')
    fig.colorbar(im, ax=ax, fraction=0.033, pad=0.03)
    #plt.savefig(f'Silhouette{e[idx]}_{s[idx]}.png')
    plt.show()

plot_sil_score(e, s, cv,)
```

Create new dataframe and write to csv to use in R

```python
#select desired columns from location dataframe and merge with relevant play information from play-by-play dataframe
r = location[['gameId', 'playId', 'Pos', 'Personnel', 'dbscan', 'kmeans']]\
.merge(pbp[['old_game_id', 'play_id', 'epa', 'play_type', 'run_location', 'pass_length', 'pass_location', 'posteam_type',\
           'game_seconds_remaining', 'score_differential',\
           'down', 'ydstogo']],\
       left_on=['gameId', 'playId'], right_on=['old_game_id', 'play_id'], how='inner')

#merge in relevant information from plays dataframe
r = r.loc[r['play_type'].isin(['pass','run'])]
r = r.merge(plays[['gameId', 'playId', 'defendersInTheBox',\
                  ]], on = ['gameId', 'playId'])
r.fillna('', inplace=True)

#classify plays to left/right as 'outside'
r.loc[r['pass_location'].isin(['left', 'right']), 'pass_location'] = 'outside'

#create column that classifies playtype by combining run/pass, play depth, and middle/outside
r['play'] = r['play_type'] + ' ' + r['run_location'] + r['pass_length'] + ' ' + r['pass_location']

print(r['play'].unique())

#remove datapoints with missing values for down and defenders in box
r = r.loc[~(r['down']=='')]
r = r.loc[~(r['defendersInTheBox']=='')]

#clean play column
r.play = r.play.apply(lambda x: x.strip())

r.rename(columns={'play': 'play_type_', 'dbscan': 'cluster_'}, inplace=True)
r = r[['cluster_', 'play_type_', 'posteam_type',\
        'game_seconds_remaining', 'defendersInTheBox',\
        'down', 'ydstogo', 'score_differential',\
        'Personnel', 'epa']]

#r=r.loc[~(r['play_type_'].isin(['run', 'pass']))]
#r.to_csv('Data/r_file.csv', index=False)

#write to csv
#r.to_csv('Data/r_file.csv')
```
